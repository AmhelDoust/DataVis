```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#first array consisting of numbers from 1 to 100
```

```python
x = np.arange(1, 101)
```

```python
#second array
```

```python
y1 = x ** 2
```

```python
#third array
```

```python
y2 = x ** 3
```

```python
print("Array x", x)
print("Array y1", y1)
print("Array y2", y2)
```

```python
plt.plot(x, y1, color='blue')  #Using plt.plot to create the line plot with x on the x-axis and y1 on the y-axis.
plt.xlabel('x')  #Adding label for x-axis
plt.ylabel('y1')  #Adding label for y-axis
plt.title('Line Plot of y1 = x^2')  #Adding title to the plot
plt.grid(True)  #Adding grid
plt.show()
```

```python
plt.scatter(x, y2, color='red') #Using plt.scatter to create a scatter plot with x on the x-axis and y2 on the y-axis.
plt.xlabel('x')  #Adding label for x-axis
plt.ylabel('y2')  #Adding label for y-axis
plt.title('Scatter Plot of y2 = x^3')  #Adding title to the plot
plt.show()  #Displaying the plot
```

```python
z = np.random.randn(1000) #Generating a random NumPy array of length 1000
```

```python
plt.figure(figsize=(8, 6))  #Adjusting figure size for better visualization
plt.hist(z, color='skyblue', edgecolor='black')  #Using plt.hist to create the histogram
plt.xlabel('Value')  #Adding label for x-axis
plt.ylabel('Frequency')  #Adding label for y-axis
plt.title('Histogram')  #Adding title to the plot
plt.grid(True)  #Adding grid to the plot for better readability
plt.show()  #Displaying the histogram
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2) #Creating a figure with two subplots

ax1.plot(x, y1, color='blue')
ax1.set_xlabel('x')
ax1.set_ylabel('y1')
ax1.set_title('Line Plot of y1 = x^2')
ax1.grid(True)

ax2.scatter(x, y2, color='red')
ax2.set_xlabel('x')
ax2.set_ylabel('y2')
ax2.set_title('Scatter Plot of y2 = x^3')
ax2.grid(True)

plt.show()
```
