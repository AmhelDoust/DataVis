```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = x ** 2
```

```python
plt.figure()
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot: y = x^2')
plt.grid(True)
plt.show()
```

```python
subjects = ['Mathematics', 'English', 'Physics', 'History', 'Art']
average_scores = [85, 92, 88, 91, 95]
```

```python
plt.figure()
```

```python
plt.figure()
plt.bar(subjects, average_scores, color='skyblue')
plt.xlabel('Subjects')
plt.ylabel('Average Scores')
plt.title('Bar Chart: Average test scores by subject')
```

```python
flavors = ['Vanilla', 'Chocolate', 'Strawberry', 'Mint', 'Other']
percentages = [25, 35, 15, 10, 15]
```

```python
plt.figure()
plt.pie(percentages, labels=flavors)
plt.title('Pie Chart: Favorite Ice Cream Flavors')
plt.axis('equal')
plt.show()
```
